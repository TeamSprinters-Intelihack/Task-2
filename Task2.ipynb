```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC

```

```python
examples = {
    'Greet': ["Hi", "How are you?", "Hello"],
    'Farewell': ["Goodbye", "See you later", "Take care"],
    'Inquiry': ["What's the weather like today?", "Can you tell me the time?", "Where is the nearest restaurant?"]}

    
```

```python
dataset = [(text, intent) for intent, texts in examples.items() for text in texts]
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform([text for text, _ in dataset])
y = [intent for _, intent in dataset]
```

```python
classifier = SVC(probability=True)
classifier.fit(X, y)


```

```python
confidence_threshold = 0.5
```

```python
def classify_intent(text):
    text_vectorized = vectorizer.transform([text])
    intent = classifier.predict(text_vectorized)[0]
    confidence = classifier.predict_proba(text_vectorized).max()
    if confidence >= confidence_threshold:
        return intent, confidence
    else:
        return "NLU fallback: Intent could not be confidently determined", confidence

```

```python
print(classify_intent("Hi"))  
```
